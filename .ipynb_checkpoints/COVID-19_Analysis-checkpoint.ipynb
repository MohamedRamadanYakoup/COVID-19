```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
from IPython.display import Javascript
from IPython.core.display import display
from IPython.core.display import HTML

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('csse_covid_19_data/csse_covid_19_daily_reports//03-16-2020.csv')
df.head()
```

```python
df.info()
```

# Data Preprocessing

**Change the name of the Province/State & Country/Region to a readable names**

```python
df.rename(columns={'Province/State':'State','Country/Region':'Country'}, inplace=True)
df.head()
```

**Clean the data from na values**

```python
df['State'].fillna('', inplace=True)
df.head()
```

```python
df['Active'] = df['Confirmed'] - df['Deaths'] - df['Recovered']
df.head()
```

```python
conf_country = df.groupby('Country')['Confirmed'].sum().sort_values(ascending=False).to_frame().reset_index()
conf_country = conf_country.head(20).sort_values(ascending=False, by='Confirmed')
conf_country.style.background_gradient(cmap='Reds')
```

```python
%matplotlib inline

ax, fig = plt.subplots(figsize=(30,10))
sns.barplot(conf_country.Country,conf_country.Confirmed)
plt.title('Confirmed Cases per Top 20 Countries')

plt.show()
```

```python
Country_cases = df.groupby('Country')['Confirmed', 'Deaths', 'Recovered','Active'].sum().sort_values(ascending=False,by='Confirmed').reset_index()
Country_cases.style.background_gradient(cmap='Reds')
```

```python
%matplotlib inline
temp = Country_cases.head(20)

ax, fig = plt.subplots(figsize=(30,10))
sns.barplot(temp.Country,temp.Deaths)
plt.title('Deaths Cases per Top Countries')

plt.show()
```
